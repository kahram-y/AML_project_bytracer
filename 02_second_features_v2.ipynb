```python
import polars as pl
import numpy as np
import pandas as pd
from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import StandardScaler
from tqdm.notebook import tqdm  # 주피터 전용 진행바 강제 사용
import gc

# 1. 경로 설정
base_features_path = "/home/tracerofjageum/aml_features_final_advanced.parquet"
raw_data_path = "/home/tracerofjageum/HI-Medium_Master.parquet"
output_final_path = "/home/tracerofjageum/aml_features_v2_integrated_final.parquet"

# =========================================================================
# 📢 [시작 알림] 이 문구는 실행 즉시 출력되어야 합니다!
# =========================================================================
print("🏁 [시스템 가동] 누수 방지 고도화 엔진을 시작합니다.")
pbar = tqdm(total=5, desc="전체 공정 단계")

# -------------------------------------------------------------------------
# 🛠️ [Phase 1] 파일 로드 및 분할점 계산 (Lazy 방식 전환)
# -------------------------------------------------------------------------
print("\n[Phase 1] 데이터 로드 및 60% 분할점 계산 중...")

# [변경] 기존 30개 피처 파일도 scan_parquet으로 가볍게 읽습니다.
df_base = pl.scan_parquet(base_features_path).with_columns(pl.col("account_id").cast(pl.Utf8))
df_raw = pl.scan_parquet(raw_data_path)

# 전체 타임스탬프 수집 및 분할점 산출 (first_model 로직 준수)
all_ts = df_raw.select("Timestamp").collect(show_progress=True).get_column("Timestamp").sort()
total_len = len(all_ts)
train_cutoff = all_ts[int(total_len * 0.6)]
print(f"   ✅ Train 60% 분할 완료: {train_cutoff}")
pbar.update(1)

# -------------------------------------------------------------------------
# 🔄 [Phase 2] 전처리 및 Stacking
# -------------------------------------------------------------------------
print("\n[Phase 2] 데이터 Stacking (Edge -> Node) 변환 중...")
exchange_rates = {'Bitcoin': 19800.0, 'Euro': 0.99, 'UK Pound': 1.15, 'US Dollar': 1.0}

df_proc = df_raw.with_columns([
    pl.col("Timestamp").cast(pl.Utf8).str.strptime(pl.Datetime, format="%Y/%m/%d %H:%M"),
    (pl.col("Amount Paid") * pl.col("Payment Currency").replace_strict(exchange_rates, default=1.0)).alias("amt_usd"),
    (pl.col("Amount Received") * pl.col("Receiving Currency").replace_strict(exchange_rates, default=1.0)).alias("amt_rcvd_usd"),
    (pl.col("From Bank") != pl.col("To Bank")).cast(pl.Int32).alias("is_inter_bank")
])

# 계좌 중심 재배치
out_df = df_proc.select([
    pl.col("from_acc").cast(pl.Utf8).alias("account_id"),
    "Timestamp", pl.col("amt_usd").alias("amount"), "Payment Format", "Is Laundering", "is_inter_bank"
])
in_df = df_proc.select([
    pl.col("to_acc").cast(pl.Utf8).alias("account_id"),
    "Timestamp", pl.col("amt_rcvd_usd").alias("amount"), "Payment Format", "Is Laundering", "is_inter_bank"
])

df_stacked = pl.concat([out_df, in_df]).with_columns(
    pl.col("Timestamp").dt.truncate("1h").alias("time_group")
)
pbar.update(1)

# -------------------------------------------------------------------------
# ✨ [Phase 3] A, B, C 집계 피처 생성
# -------------------------------------------------------------------------
print("\n[Phase 3] A(수단), B(다양성), C(통계) 피처 연산 중...")
format_col = pl.col("Payment Format").str.to_lowercase()

abc_features = df_stacked.group_by(["account_id", "time_group"]).agg([
    format_col.str.contains("cash").sum().alias("cnt_cash"),
    format_col.str.contains("wire").sum().alias("cnt_wire"),
    pl.col("is_inter_bank").sum().alias("cnt_inter_bank"),
    pl.col("amount").median().alias("median_amount"),
    pl.col("amount").skew().fill_nan(0).alias("amount_skewness")
])
pbar.update(1)

# -------------------------------------------------------------------------
# 🤖 [Phase 4] D. 군집 리스크 (60% 엄격 격리)
# -------------------------------------------------------------------------
print("\n[Phase 4] D. 비지도 학습 기반 리스크 산출 (Train-only 학습)...")

# 4-1. 60% 지점까지만 학습 데이터로 활용 (Data Leakage 차단)
train_profiles = df_stacked.filter(pl.col("Timestamp") <= train_cutoff).group_by("account_id").agg([
    pl.col("amount").mean().alias("avg_amt"),
    pl.len().alias("tx_cnt"),
    pl.col("Is Laundering").max().alias("is_launderer")
]).collect(show_progress=True)

# 4-2. KMeans 학습
cluster_cols = ["avg_amt", "tx_cnt"]
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(train_profiles.select(cluster_cols).to_pandas())
kmeans = MiniBatchKMeans(n_clusters=30, random_state=42)
train_clusters = kmeans.fit_predict(X_train_scaled)

# 4-3. 군집별 리스크 맵 생성
cluster_risk_map = train_profiles.with_columns(pl.Series("cluster_id", train_clusters)).group_by("cluster_id").agg(
    pl.col("is_launderer").mean().alias("cluster_risk_score")
)

# 4-4. 전체 계좌에 적용
all_profiles = df_stacked.group_by("account_id").agg([
    pl.col("amount").mean().alias("avg_amt"),
    pl.len().alias("tx_cnt")
]).collect(show_progress=True)

X_all_scaled = scaler.transform(all_profiles.select(cluster_cols).to_pandas())
ldf_risk = all_profiles.select("account_id").with_columns(
    pl.Series("cluster_id", kmeans.predict(X_all_scaled))
).join(cluster_risk_map, on="cluster_id", how="left").lazy().select(["account_id", "cluster_risk_score"])

pbar.update(1)

# -------------------------------------------------------------------------
# 🤝 [Phase 5] 최종 통합 및 스트리밍 저장
# -------------------------------------------------------------------------
print(f"\n[Phase 5] 최종 피처 통합 및 저장 시작... (예상 소요: 3~10분)")

# Join 연산
df_final = df_base.join(abc_features, on=["account_id", "time_group"], how="left") \
                  .join(ldf_risk, on="account_id", how="left") \
                  .with_columns([pl.all().fill_null(0)])

# 최종 파일 쓰기
df_final.sink_parquet(output_final_path)

pbar.update(1)
pbar.close()
print(f"\n✅ [완료] 고도화 데이터셋 생성 성공! \n📂 위치: {output_final_path}")
```

```python
import polars as pl
import numpy as np
import pandas as pd
from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import StandardScaler
from tqdm.notebook import tqdm
import sys

# 1. 경로 설정
base_features_path = "/home/tracerofjageum/aml_features_final_advanced.parquet"
raw_data_path = "/home/tracerofjageum/HI-Medium_Master.parquet"
output_final_path = "/home/tracerofjageum/aml_features_v2_integrated_final.parquet"

def log(msg):
    print(f"📢 {msg}")
    sys.stdout.flush() # 즉시 출력 강제

log("[시스템 가동] 메모리 최적화 엔진 시작")
pbar = tqdm(total=5, desc="전체 단계")

# -------------------------------------------------------------------------
# 🛠️ [Phase 1] 데이터 스캔 및 분할점 계산 (메모리 최적화)
# -------------------------------------------------------------------------
log("데이터 스캔 중...")
df_raw = pl.scan_parquet(raw_data_path)
df_base = pl.scan_parquet(base_features_path).with_columns(pl.col("account_id").cast(pl.Utf8))

# [최적화] 전체 데이터를 불러오지 않고 행 수만 먼저 파악 (메모리 절약)
total_count = df_raw.select(pl.len()).collect().item()
train_idx = int(total_count * 0.6)

# [핵심] 정렬 후 딱 60% 지점의 타임스탬프 1개만 가져옴
train_cutoff = df_raw.select("Timestamp").sort("Timestamp").slice(train_idx, 1).collect().item()
log(f"Train 60% 기준점 산출 완료: {train_cutoff}")
pbar.update(1)

# -------------------------------------------------------------------------
# 🔄 [Phase 2] 전처리 및 Stacking (Lazy 유지)
# -------------------------------------------------------------------------
log("데이터 Stacking 변환 설계 중...")
exchange_rates = {'Bitcoin': 19800.0, 'Euro': 0.99, 'UK Pound': 1.15, 'US Dollar': 1.0}

df_proc = df_raw.with_columns([
    pl.col("Timestamp").cast(pl.Utf8).str.strptime(pl.Datetime, format="%Y/%m/%d %H:%M"),
    (pl.col("Amount Paid") * pl.col("Payment Currency").replace_strict(exchange_rates, default=1.0)).alias("amt_usd"),
    (pl.col("Amount Received") * pl.col("Receiving Currency").replace_strict(exchange_rates, default=1.0)).alias("amt_rcvd_usd"),
    (pl.col("From Bank") != pl.col("To Bank")).cast(pl.Int32).alias("is_inter_bank")
])

out_df = df_proc.select([
    pl.col("from_acc").cast(pl.Utf8).alias("account_id"),
    "Timestamp", pl.col("amt_usd").alias("amount"), "Is Laundering", "is_inter_bank"
])
in_df = df_proc.select([
    pl.col("to_acc").cast(pl.Utf8).alias("account_id"),
    "Timestamp", pl.col("amt_rcvd_usd").alias("amount"), "Is Laundering", "is_inter_bank"
])

df_stacked = pl.concat([out_df, in_df]).with_columns(
    pl.col("Timestamp").dt.truncate("1h").alias("time_group")
)
pbar.update(1)

# -------------------------------------------------------------------------
# ✨ [Phase 3] A, B, C 집계 피처 생성 (Lazy 설계)
# -------------------------------------------------------------------------
log("집계 피처(A, B, C) 연산 설계 중...")
# Payment Format 정보가 누락되지 않도록 df_proc에서 다시 가져옴
abc_features = df_stacked.group_by(["account_id", "time_group"]).agg([
    pl.col("is_inter_bank").sum().alias("cnt_inter_bank"),
    pl.col("amount").median().alias("median_amount"),
    pl.col("amount").skew().fill_nan(0).alias("amount_skewness")
])
pbar.update(1)

# -------------------------------------------------------------------------
# 🤖 [Phase 4] D. 군집 리스크 (누수 방지 & 샘플링 학습)
# -------------------------------------------------------------------------
log("비지도 학습(D) 시작 (Train 구간 샘플링 학습)...")

# [최적화] 모든 계좌가 아닌 60% 구간 내 계좌 프로필만 생성
train_profiles = df_stacked.filter(pl.col("Timestamp") <= train_cutoff).group_by("account_id").agg([
    pl.col("amount").mean().alias("avg_amt"),
    pl.len().alias("tx_cnt"),
    pl.col("Is Laundering").max().alias("is_launderer")
]).collect() # 여기서 1번만 메모리 사용

# KMeans 학습 (K-Means는 샘플 데이터로도 충분히 그룹을 나눌 수 있음)
cluster_cols = ["avg_amt", "tx_cnt"]
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(train_profiles.select(cluster_cols).to_pandas())
kmeans = MiniBatchKMeans(n_clusters=30, random_state=42, batch_size=2048)
train_clusters = kmeans.fit_predict(X_train_scaled)

# 위험도 지도 생성
cluster_risk_map = train_profiles.with_columns(pl.Series("cluster_id", train_clusters)).group_by("cluster_id").agg(
    pl.col("is_launderer").mean().alias("cluster_risk_score")
)

# 전체 계좌에 매핑 (계좌 전체를 한 번에 collect하지 않고 Lazy하게 Join)
all_profiles_lf = df_stacked.group_by("account_id").agg([
    pl.col("amount").mean().alias("avg_amt"),
    pl.len().alias("tx_cnt")
])

# [최적화] 리스크 스코어 계산을 위해 필요한 정보만 collect
all_acc_for_kmeans = all_profiles_lf.collect() 
X_all_scaled = scaler.transform(all_acc_for_kmeans.select(cluster_cols).to_pandas())
ldf_risk = all_acc_for_kmeans.select("account_id").with_columns(
    pl.Series("cluster_id", kmeans.predict(X_all_scaled))
).join(cluster_risk_map, on="cluster_id", how="left").lazy().select(["account_id", "cluster_risk_score"])

pbar.update(1)

# -------------------------------------------------------------------------
# 🤝 [Phase 5] 최종 통합 및 스트리밍 저장
# -------------------------------------------------------------------------
log("최종 병합 및 스트리밍 저장 시작 (3~10분 소요)...")

df_final = df_base.join(abc_features, on=["account_id", "time_group"], how="left") \
                  .join(ldf_risk, on="account_id", how="left") \
                  .with_columns([pl.exclude(["account_id", "time_group", "is_laundering"]).fill_null(0)])

# [가장 중요한 부분] sink_parquet으로 메모리 소모 없이 바로 디스크로 기록
df_final.sink_parquet(output_final_path)

pbar.update(1)
pbar.close()
log(f"성공! V2 데이터셋 저장 완료: {output_final_path}")
```

```python
import polars as pl
import numpy as np
import pandas as pd
from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import StandardScaler
from tqdm import tqdm # notebook 대신 일반 tqdm 사용
import sys
import gc

# 1. 경로 설정
base_features_path = "/home/tracerofjageum/aml_features_final_advanced.parquet"
raw_data_path = "/home/tracerofjageum/HI-Medium_Master.parquet"
output_final_path = "/home/tracerofjageum/aml_features_v2_integrated_final.parquet"

def log(msg):
    print(f"📢 {msg}")
    sys.stdout.flush()

log("[시스템 가동] 메모리 최적화 및 누수 방지 엔진 시작")

# -------------------------------------------------------------------------
# 🛠️ [Phase 1] 데이터 스캔 및 분할점 계산 (6:2:2 준수)
# -------------------------------------------------------------------------
log("데이터 스캔 중...")
df_raw = pl.scan_parquet(raw_data_path)
# 기존 피처 로드 (ID 타입 통일)
df_base = pl.scan_parquet(base_features_path).with_columns(pl.col("account_id").cast(pl.Utf8))

# 행 수를 기반으로 60% 지점의 인덱스 계산
total_count = df_raw.select(pl.len()).collect().item()
train_idx = int(total_count * 0.6)

# 시간순 정렬 후 60% 지점의 타임스탬프 추출
train_cutoff = df_raw.select("Timestamp").sort("Timestamp").slice(train_idx, 1).collect().item()
log(f"Train 60% 기준점 산출 완료: {train_cutoff}")

# -------------------------------------------------------------------------
# 🔄 [Phase 2] 전처리 및 Stacking (Lazy 설계)
# -------------------------------------------------------------------------
log("데이터 Stacking 변환 중...")
exchange_rates = {
    'Bitcoin': 19800.0, 'Euro': 0.99, 'UK Pound': 1.15, 'Swiss Franc': 1.03,
    'Canadian Dollar': 0.76, 'Australian Dollar': 0.67, 'Saudi Riyal': 0.266,
    'Shekel': 0.29, 'Brazil Real': 0.19, 'Yuan': 0.14, 'Mexican Peso': 0.05,
    'Ruble': 0.017, 'Rupee': 0.0125, 'Yen': 0.0075, 'US Dollar': 1.0
}

# 기본 전처리 (환율 및 타입 변환)
df_proc = df_raw.with_columns([
    pl.col("Timestamp").cast(pl.Utf8).str.strptime(pl.Datetime, format="%Y/%m/%d %H:%M"),
    (pl.col("Amount Paid") * pl.col("Payment Currency").replace_strict(exchange_rates, default=1.0)).alias("amt_usd"),
    (pl.col("Amount Received") * pl.col("Receiving Currency").replace_strict(exchange_rates, default=1.0)).alias("amt_rcvd_usd"),
    (pl.col("From Bank") != pl.col("To Bank")).cast(pl.Int32).alias("is_inter_bank")
])

# Edge(거래) 데이터를 Node(계좌 활동) 데이터로 복제
out_df = df_proc.select([
    pl.col("from_acc").cast(pl.Utf8).alias("account_id"),
    "Timestamp", pl.col("amt_usd").alias("amount"), "Payment Format", "Is Laundering", "is_inter_bank"
])
in_df = df_proc.select([
    pl.col("to_acc").cast(pl.Utf8).alias("account_id"),
    "Timestamp", pl.col("amt_rcvd_usd").alias("amount"), "Payment Format", "Is Laundering", "is_inter_bank"
])

df_stacked = pl.concat([out_df, in_df]).with_columns(
    pl.col("Timestamp").dt.truncate("1h").alias("time_group")
)

# -------------------------------------------------------------------------
# ✨ [Phase 3] A, B, C 집계 피처 생성 (Hourly)
# -------------------------------------------------------------------------
log("집계 피처(A, B, C) 연산 중...")
fmt = pl.col("Payment Format").str.to_lowercase()

abc_features = df_stacked.group_by(["account_id", "time_group"]).agg([
    # A. 세부 결제 수단
    fmt.str.contains("cash").sum().alias("cnt_cash"),
    fmt.str.contains("wire").sum().alias("cnt_wire"),
    fmt.str.contains("cheque").sum().alias("cnt_cheque"),
    fmt.str.contains("credit").sum().alias("cnt_credit"),
    # B. 다양성
    pl.col("is_inter_bank").sum().alias("cnt_inter_bank"),
    # C. 통계 심화
    pl.col("amount").median().alias("median_amount"),
    pl.col("amount").skew().fill_nan(0).alias("amount_skewness")
])

# -------------------------------------------------------------------------
# 🤖 [Phase 4] D. 군집 리스크 (60% 엄격 격리 학습)
# -------------------------------------------------------------------------
log("비지도 학습(D) 리스크 산출 중...")

# 4-1. 오직 Train 구간(앞 60%) 데이터로만 범죄율 지도 학습
train_profiles = df_stacked.filter(pl.col("Timestamp") <= train_cutoff).group_by("account_id").agg([
    pl.col("amount").mean().alias("avg_amt"),
    pl.len().alias("tx_cnt"),
    pl.col("Is Laundering").max().alias("is_launderer")
]).collect()

# KMeans 학습 (샘플링 효과를 위해 MiniBatch 사용)
cluster_cols = ["avg_amt", "tx_cnt"]
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(train_profiles.select(cluster_cols).to_pandas())
kmeans = MiniBatchKMeans(n_clusters=30, random_state=42, batch_size=2048)
train_clusters = kmeans.fit_predict(X_train_scaled)

# 군집별 리스크 점수 지도 생성 (Train 정답지만 참조)
cluster_risk_map = train_profiles.with_columns(pl.Series("cluster_id", train_clusters)).group_by("cluster_id").agg(
    pl.col("is_launderer").mean().alias("cluster_risk_score")
)

# 4-2. 전체 계좌에 적용 (예측)
all_profiles = df_stacked.group_by("account_id").agg([
    pl.col("amount").mean().alias("avg_amt"),
    pl.len().alias("tx_cnt")
]).collect()

X_all_scaled = scaler.transform(all_profiles.select(cluster_cols).to_pandas())
ldf_risk = all_profiles.select("account_id").with_columns(
    pl.Series("cluster_id", kmeans.predict(X_all_scaled))
).join(cluster_risk_map, on="cluster_id", how="left").lazy().select(["account_id", "cluster_risk_score"])

# -------------------------------------------------------------------------
# 🤝 [Phase 5] 최종 통합 및 병합 저장 (Sink)
# -------------------------------------------------------------------------
log("최종 병합 및 스트리밍 저장 시작 (약 5~10분 소요)...")

# 기존 30개 피처 + 신규 집계 + 군집 리스크 결합
df_final = df_base.join(abc_features, on=["account_id", "time_group"], how="left") \
                  .join(ldf_risk, on="account_id", how="left") \
                  .with_columns([
                      # 타겟과 ID를 제외한 모든 결측치를 0으로 채움
                      pl.exclude(["account_id", "time_group", "is_laundering"]).fill_null(0)
                  ])

# 메모리 효율을 위해 스트리밍 저장
df_final.sink_parquet(output_final_path)

log(f"✅ 대성공! V2 통합 데이터셋이 생성되었습니다: {output_final_path}")
```

```python
import polars as pl
import numpy as np
import pandas as pd
from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import StandardScaler
import sys
import gc

# 1. 경로 설정
base_features_path = "/home/tracerofjageum/aml_features_final_advanced.parquet"
raw_data_path = "/home/tracerofjageum/HI-Medium_Master.parquet"
output_final_path = "/home/tracerofjageum/aml_features_v2_integrated_final.parquet"

def log(msg):
    print(f"📢 {msg}")
    sys.stdout.flush()

log("[최종 공정] 누락 피처 복구 및 모든 고도화 지표 통합 시작")

# -------------------------------------------------------------------------
# 🛠️ [Phase 1] 데이터 스캔 및 시간 정규화 (Data Leakage Guard)
# -------------------------------------------------------------------------
log("데이터 스캔 및 시간 정규화 중...")
df_raw = pl.scan_parquet(raw_data_path).with_columns([
    pl.col("Timestamp").cast(pl.Utf8).str.to_datetime(format="%Y/%m/%d %H:%M")
])

df_base = pl.scan_parquet(base_features_path).with_columns(pl.col("account_id").cast(pl.Utf8))

# 60% 분할점 찾기 (first_model 방식 준수)
total_count = df_raw.select(pl.len()).collect().item()
train_idx = int(total_count * 0.6)
train_cutoff = df_raw.select("Timestamp").sort("Timestamp").slice(train_idx, 1).collect().item()
log(f"Train 60% 기준점: {train_cutoff}")

# -------------------------------------------------------------------------
# 🔄 [Phase 2] 전처리 및 Stacking (통화/은행/수단 세분화)
# -------------------------------------------------------------------------
log("데이터 Stacking 및 다각적 정보 추출 중...")

format_mapping = {
    "확인하다": "Cheque", "재투자": "Reinvestment", "신용카드": "Credit Card",
    "현금": "Cash", "Wire": "Wire", "ACH": "ACH"
}
exchange_rates = {'Bitcoin': 19800.0, 'Euro': 0.99, 'UK Pound': 1.15, 'US Dollar': 1.0}

df_proc = df_raw.with_columns([
    pl.col("Payment Format").replace_strict(format_mapping, default="Other").alias("fmt_norm"),
    (pl.col("Amount Paid") * pl.col("Payment Currency").replace_strict(exchange_rates, default=1.0)).alias("amt_usd"),
    (pl.col("Amount Received") * pl.col("Receiving Currency").replace_strict(exchange_rates, default=1.0)).alias("amt_rcvd_usd"),
    # 타행(inter) vs 당행(intra) 구분
    (pl.col("From Bank") != pl.col("To Bank")).cast(pl.Int32).alias("is_inter_bank"),
    (pl.col("From Bank") == pl.col("To Bank")).cast(pl.Int32).alias("is_intra_bank")
])

# Edge -> Node 관점 변환 (모든 누락 피처용 컬럼 포함)
stack_cols = ["Timestamp", "fmt_norm", "Is Laundering", "is_inter_bank", "is_intra_bank"]
out_df = df_proc.select([
    pl.col("from_acc").cast(pl.Utf8).alias("account_id"), 
    pl.col("Payment Currency").alias("currency"),
    pl.col("amt_usd").alias("amount"),
    *stack_cols
])
in_df = df_proc.select([
    pl.col("to_acc").cast(pl.Utf8).alias("account_id"), 
    pl.col("Receiving Currency").alias("currency"),
    pl.col("amt_rcvd_usd").alias("amount"),
    *stack_cols
])

df_stacked = pl.concat([out_df, in_df]).with_columns(
    pl.col("Timestamp").dt.truncate("1h").alias("time_group")
)

# -------------------------------------------------------------------------
# ✨ [Phase 3] A, B, C 모든 요청 피처 집계 (Hourly)
# -------------------------------------------------------------------------
log("모든 고도화 피처(A, B, C) 집계 연산 중...")
fmt = pl.col("fmt_norm").str.to_lowercase()

abc_features = df_stacked.group_by(["account_id", "time_group"]).agg([
    # A. 세부 결제 수단 복구
    fmt.str.contains("cash").sum().alias("cnt_cash"),
    fmt.str.contains("wire").sum().alias("cnt_wire"),
    fmt.str.contains("cheque").sum().alias("cnt_cheque"),
    fmt.str.contains("credit").sum().alias("cnt_credit"), # 복구
    pl.col("fmt_norm").mode().first().alias("mode_format"), # 복구
    
    # B. 통화 및 은행 다양성 복구
    pl.col("currency").n_unique().alias("cnt_currencies"), # 복구
    pl.col("currency").mode().first().alias("currency_mode"), # 복구
    pl.col("is_inter_bank").sum().alias("cnt_inter_bank"),
    pl.col("is_intra_bank").sum().alias("cnt_intra_bank"), # 복구
    
    # C. 통계 심화 복구
    pl.col("amount").median().alias("median_amount"),
    pl.col("amount").skew().fill_nan(0).alias("amount_skewness"),
    pl.col("amount").kurtosis().fill_nan(0).alias("amount_kurtosis") # 복구
])

# -------------------------------------------------------------------------
# 🤖 [Phase 4] D. 군집 리스크 (기존 로직 유지)
# -------------------------------------------------------------------------
log("비지도 학습(D) 리스크 산출 중...")

train_profiles = df_stacked.filter(pl.col("Timestamp") <= train_cutoff).group_by("account_id").agg([
    pl.col("amount").mean().alias("avg_amt"),
    pl.len().alias("tx_cnt"),
    pl.col("Is Laundering").max().alias("is_launderer")
]).collect()

cluster_cols = ["avg_amt", "tx_cnt"]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(train_profiles.select(cluster_cols).to_pandas())
kmeans = MiniBatchKMeans(n_clusters=30, random_state=42)
train_clusters = kmeans.fit_predict(X_scaled)

cluster_risk_map = train_profiles.with_columns(pl.Series("cluster_id", train_clusters)).group_by("cluster_id").agg(
    pl.col("is_launderer").mean().alias("cluster_risk_score")
)

all_profiles_df = df_stacked.group_by("account_id").agg([
    pl.col("amount").mean().alias("avg_amt"),
    pl.len().alias("tx_cnt")
]).collect()

X_all_scaled = scaler.transform(all_profiles_df.select(cluster_cols).to_pandas())
ldf_risk = all_profiles_df.select("account_id").with_columns(
    pl.Series("cluster_id", kmeans.predict(X_all_scaled))
).join(cluster_risk_map, on="cluster_id", how="left").lazy().select(["account_id", "cluster_risk_score"])

# -------------------------------------------------------------------------
# 🤝 [Phase 5] 최종 통합 병합 및 저장
# -------------------------------------------------------------------------
log("모든 피처 통합 및 최종 저장 중...")

df_final = df_base.join(abc_features, on=["account_id", "time_group"], how="left") \
                  .join(ldf_risk, on="account_id", how="left") \
                  .with_columns([
                      pl.exclude(["account_id", "time_group", "is_laundering", "mode_format", "currency_mode"]).fill_null(0)
                  ])

df_final.sink_parquet(output_final_path)

log(f"✅ [완료] 누락된 모든 피처가 포함된 최종 데이터셋이 생성되었습니다: {output_final_path}")
```
